```python

def Euclediandistance(x,y):
    return np.linalg.norm(x-y)

```

```python
def findCentroids(clusters):
    return [np.mean(cluster, axis=0) for cluster in clusters]
```

```python


def Centroidscompare(p1, p2):
    if len(p1) != len(p2):
        return False
    for p1, p2 in zip(p1, p2):
        if p1[0] != p2[0] or p1[1] != p2[1]:
            return False   
    return True
```

```python
def closestCentroid(point, centroids):
    min_distance = float('inf')
    clusterPoint = None
    for i,centroid in enumerate(centroids):
        dist = Euclediandistance(point, centroid)
        if dist < min_distance:
            min_distance = dist
            clusterPoint = i
    return clusterPoint


def MyKMeans(k, centroids, points):
    clusters = [[] for _ in range(k)]
    for point in points:
        clusters[closestCentroid(point, centroids)].append(point)
        
    new_centroids = findCentroids(clusters)
    
    if not Centroidscompare(centroids, new_centroids):
        clusters = MyKMeans(k, new_centroids, points)
        
    return array(new_centroids)
```

```python
k = 12
centroids = X[:k]               #picking the first k points as centroids
centersMyKmeans = MyKMeans(k, centroids, X)
centersMyKmeans
```
